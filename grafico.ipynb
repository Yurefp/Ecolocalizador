```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sn
import numpy as np

tab = pd.read_csv('C:tabela.csv')
tab.head()
```

```python
#tab.info()
#tab.describe()
tab.Preciso.value_counts()
```

```python
#Para precisão:

tab_pr = tab.drop(columns=['Preciso', 'Acuracia', 'R.A.N.', 'R.A.A.', 'R.A.N.-R.A.A.', 'Acurado'])
tab_pr.head()

#Heatmap dos dados de precisão
correlacao = tab_pr.corr()
sn.set(rc = {'figure.figsize':(10,7)})
plt.figure()
sn.heatmap(data=correlacao, annot=True)
```

```python
#Para acuracia:

tab_ac = tab.drop(columns=['Preciso', 'Precisao', 'R.P.N.', 'R.P.A.', 'R.P.N.-R.P.A.', 'Acurado'])
tab_ac.head()

#Heatmap dos dados de precisão
correlacao = tab_ac.corr()
sn.set(rc = {'figure.figsize':(10,7)})
plt.figure()
sn.heatmap(data=correlacao, annot=True)
```
